**Matplot lib using NUMPY**

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
plt.style.use('classic')
```

```python
x=np.linspace(0,10,100)
print(x)
```

```python
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

```python
plt.figure()
plt.subplot(2,1,1)
plt.plot(x,np.sin(x))
plt.subplot(2,1,2)
plt.plot(x,np.cos(x))
```

```python
plt.figure()
plt.subplot(2,2,1)
plt.plot(x,np.sin(x))
plt.subplot(2,2,2)
plt.plot(x,np.cos(x))
plt.subplot(2,2,3)
plt.plot(x,np.tan(x))
```

```python
plt.figure(figsize=(15,4))
plt.subplot(2,2,1)
plt.plot(x,np.sin(x))
plt.subplot(2,2,2)
plt.plot(x,np.cos(x))
plt.subplot(2,2,3)
plt.plot(x,np.tan(x))
```
